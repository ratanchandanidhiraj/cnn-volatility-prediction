```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense
from sklearn.metrics import mean_squared_error, mean_absolute_error
# Set random seed
np.random.seed(42)
tf.random.set_seed(42)

# STEP 1: Download Data



print("=" * 50)
print("STEP 1: Downloading data...")
print("=" * 50)
TICKERS = ["AAPL", "MSFT", "GOOGL"]
START = "2022-01-01"
END = "2024-01-01"
dfs = {}
for ticker in TICKERS:
 print(f"Downloading {ticker}...")
 df = yf.download(ticker, START, END, progress=False)
 df = df.dropna()
 dfs[ticker] = df
 print(f" {ticker}: {len(df)} days")

# STEP 2: Calculate Volatility


print("\n" + "=" * 50)
print("STEP 2: Calculating volatility...")
print("=" * 50)
for ticker in TICKERS:
 df = dfs[ticker]

 # Returns
 df['return'] = np.log(df['Close'] / df['Close'].shift(1))

 # Parkinson volatility
 df['volatility'] = (np.log(df['High'] / df['Low'])**2) / (4 *
np.log(2))


 df = df.dropna()
 dfs[ticker] = df
 print(f"{ticker}: Volatility range [{df['volatility'].min():.6f}, {df['volatility'].max():.6f}]")

# STEP 3: Create Sliding Windows

print("\n" + "=" * 50)
print("STEP 3: Creating training examples...")
print("=" * 50)
WINDOW = 10
FEATURES = ['Open', 'High', 'Low', 'Close', 'Volume', 'volatility']
X_list = []
y_list = []
for ticker in TICKERS:
 df = dfs[ticker]

 for i in range(WINDOW, len(df) - 1):
  X_list.append(df[FEATURES].iloc[i-WINDOW:i].values)
  y_list.append(df['volatility'].iloc[i+1])
X = np.array(X_list)
y = np.array(y_list)
print(f"Total examples: {len(X)}")
print(f"Input shape: {X.shape}")
print(f"Output shape: {y.shape}")


# STEP 4: Split Data
print("\n" + "=" * 50)
print("STEP 4: Splitting data...")
print("=" * 50)
n_train = int(len(X) * 0.70)
n_val = int(len(X) * 0.15)
X_train = X[:n_train]
X_val = X[n_train:n_train+n_val]
X_test = X[n_train+n_val:]
y_train = y[:n_train]
y_val = y[n_train:n_train+n_val]
y_test = y[n_train+n_val:]
print(f"Training: {len(X_train)} examples")
print(f"Validation: {len(X_val)} examples")
print(f"Test: {len(X_test)} examples")


# STEP 5: Normalize

print("\n" + "=" * 50)
print("STEP 5: Normalizing data...")
print("=" * 50)
mean = X_train.mean(axis=(0,1))
std = X_train.std(axis=(0,1))
X_train_norm = (X_train - mean) / std
X_val_norm = (X_val - mean) / std
X_test_norm = (X_test - mean) / std
print("Normalization complete!")


# STEP 6: Build Model

print("\n" + "=" * 50)
print("STEP 6: Building model...")
print("=" * 50)
model = Sequential([
 Conv1D(32, kernel_size=3, activation='relu', input_shape=(WINDOW,
len(FEATURES))),
 MaxPooling1D(pool_size=2),
 Conv1D(64, kernel_size=3, activation='relu'),
 Flatten(),
 Dense(32, activation='relu'),
 Dense(1)
])
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
model.summary()


# STEP 7: Train Model

print("\n" + "=" * 50)
print("STEP 7: Training model...")
print("=" * 50)
history = model.fit(
 X_train_norm, y_train,
 validation_data=(X_val_norm, y_val),
 epochs=30,
 batch_size=32,
 verbose=1
)
# Plot training progress
plt.figure(figsize=(10, 4))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Training Progress')
plt.xlabel('Epoch')
plt.ylabel('Loss (MSE)')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('training_progress.png', dpi=150)
plt.show()



# STEP 8: Evaluate Model


print("\n" + "=" * 50)
print("STEP 8: Evaluating model...")
print("=" * 50)
predictions = model.predict(X_test_norm).flatten()
mse = mean_squared_error(y_test, predictions)
mae = mean_absolute_error(y_test, predictions)
rmse = np.sqrt(mse)
print(f"\nModel Performance:")
print(f" RMSE: {rmse:.6f}")
print(f" MAE: {mae:.6f}")
# Naive baseline
naive_predictions = y_test[:-1]
naive_actual = y_test[1:]
naive_rmse = np.sqrt(mean_squared_error(naive_actual, naive_predictions))
print(f"\nNaive Baseline RMSE: {naive_rmse:.6f}")
print(f"Improvement: {((naive_rmse - rmse) / naive_rmse * 100):.1f}%")



# STEP 9: Visualize Results

print("\n" + "=" * 50)
print("STEP 9: Creating visualizations...")
print("=" * 50)
# Plot 1: Predictions over time
plt.figure(figsize=(12, 5))
plt.plot(y_test[:100], label='Actual', marker='o', markersize=3, alpha=0.7)
plt.plot(predictions[:100], label='Predicted', marker='x', markersize=3,
alpha=0.7)
plt.title('Volatility Predictions vs Actual (First 100 Test Samples)')
plt.xlabel('Test Sample')
plt.ylabel('Volatility')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('predictions_timeline.png', dpi=150)
plt.show()
# Plot 2: Scatter plot
plt.figure(figsize=(6, 6))
plt.scatter(y_test, predictions, alpha=0.5)
plt.plot([y_test.min(), y_test.max()],
 [y_test.min(), y_test.max()],
 'r--', linewidth=2, label='Perfect Prediction')
plt.xlabel('Actual Volatility')
plt.ylabel('Predicted Volatility')
plt.title('Prediction Accuracy')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('prediction_scatter.png', dpi=150)
plt.show()
print("\n" + "=" * 50)
print("COMPLETE! Check your plots and results.")
print("=" * 50)
```

```python
#3
# Volatility Prediction using CNN

import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# STEP 1: Download Stock Data
print("=" * 50)
print("STEP 1: Downloading data...")
print("=" * 50)

TICKERS = ["AAPL", "MSFT", "GOOGL"]
START = "2022-01-01"
END = "2024-01-01"

dfs = {}

for ticker in TICKERS:
    print(f"Downloading {ticker}...")
    df = yf.download(ticker, START, END, progress=False)
    df = df.dropna()
    dfs[ticker] = df
    print(f"{ticker}: {len(df)} days")

# STEP 2: Feature Engineering + Volatility
print("\n" + "=" * 50)
print("STEP 2: Calculating volatility and features...")
print("=" * 50)

for ticker in TICKERS:
    df = dfs[ticker]

    # Log returns
    df['return'] = np.log(df['Close'] / df['Close'].shift(1))

    # Parkinson volatility
    df['volatility'] = (np.log(df['High'] / df['Low']) ** 2) / (4 * np.log(2))

    # Rolling statistics (OPTIONAL EXTENSION)
    df['rolling_mean_5'] = df['return'].rolling(5).mean()
    df['rolling_std_5'] = df['return'].rolling(5).std()

    df = df.dropna()
    dfs[ticker] = df

    print(f"{ticker}: Volatility range [{df['volatility'].min():.6f}, {df['volatility'].max():.6f}]")

# STEP 3: Create Sliding Windows

print("\n" + "=" * 50)
print("STEP 3: Creating training examples...")
print("=" * 50)

WINDOW = 20  # OPTIONAL EXTENSION (changed from 10)

FEATURES = [
    'Open', 'High', 'Low', 'Close', 'Volume',
    'return', 'volatility',
    'rolling_mean_5', 'rolling_std_5'
]

X_list, y_list = [], []

for ticker in TICKERS:
    df = dfs[ticker]

    for i in range(WINDOW, len(df) - 1):
        X_list.append(df[FEATURES].iloc[i-WINDOW:i].values)
        y_list.append(df['volatility'].iloc[i + 1])

X = np.array(X_list)
y = np.array(y_list)

print(f"Total samples: {len(X)}")
print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}")


# STEP 4: Train / Validation / Test Split

print("\n" + "=" * 50)
print("STEP 4: Splitting data...")
print("=" * 50)

n_train = int(len(X) * 0.70)
n_val = int(len(X) * 0.15)

X_train = X[:n_train]
X_val = X[n_train:n_train + n_val]
X_test = X[n_train + n_val:]

y_train = y[:n_train]
y_val = y[n_train:n_train + n_val]
y_test = y[n_train + n_val:]

print(f"Train: {len(X_train)}")
print(f"Validation: {len(X_val)}")
print(f"Test: {len(X_test)}")


# STEP 5: Normalize X and y

print("\n" + "=" * 50)
print("STEP 5: Normalizing data...")
print("=" * 50)

# Normalize X
X_mean = X_train.mean(axis=(0, 1))
X_std = X_train.std(axis=(0, 1))

X_train_norm = (X_train - X_mean) / X_std
X_val_norm = (X_val - X_mean) / X_std
X_test_norm = (X_test - X_mean) / X_std

# Normalize y (IMPORTANT FIX)
y_mean = y_train.mean()
y_std = y_train.std()

y_train_norm = (y_train - y_mean) / y_std
y_val_norm = (y_val - y_mean) / y_std

print("Normalization completed.")


# STEP 6: Build CNN Model (Improved Depth)

print("\n" + "=" * 50)
print("STEP 6: Building CNN model...")
print("=" * 50)

model = Sequential([
    Conv1D(32, kernel_size=5, activation='relu', input_shape=(WINDOW, len(FEATURES))),
    MaxPooling1D(2),
    Conv1D(64, kernel_size=3, activation='relu'),
    Conv1D(64, kernel_size=3, activation='relu'),
    Flatten(),
    Dense(32, activation='relu'),
    Dense(1)
])

model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae']
)

model.summary()


# STEP 7: Train Model

print("\n" + "=" * 50)
print("STEP 7: Training model...")
print("=" * 50)

history = model.fit(
    X_train_norm, y_train_norm,
    validation_data=(X_val_norm, y_val_norm),
    epochs=30,
    batch_size=32,
    verbose=1
)


# STEP 8: Evaluate Model (De-normalized)

print("\n" + "=" * 50)
print("STEP 8: Evaluating model...")
print("=" * 50)

# Predict (normalized)
y_pred_norm = model.predict(X_test_norm).flatten()

# De-normalize predictions
y_pred = y_pred_norm * y_std + y_mean

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)

print("\nModel Performance:")
print(f"RMSE: {rmse:.6f}")
print(f"MAE:  {mae:.6f}")


# STEP 9: Naive Baseline Comparison

naive_pred = y_test[:-1]
naive_actual = y_test[1:]

naive_rmse = np.sqrt(mean_squared_error(naive_actual, naive_pred))

print("\nNaive Baseline RMSE:", round(naive_rmse, 6))
print("Improvement (%):", round((naive_rmse - rmse) / naive_rmse * 100, 2))


# STEP 10: Visualizations

print("\n" + "=" * 50)
print("STEP 10: Visualizing results...")
print("=" * 50)

# Timeline plot
plt.figure(figsize=(12, 5))
plt.plot(y_test[:100], label='Actual', marker='o', markersize=3)
plt.plot(y_pred[:100], label='Predicted', marker='x', markersize=3)
plt.title('Volatility Prediction vs Actual')
plt.xlabel('Test Sample')
plt.ylabel('Volatility')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# Scatter plot
plt.figure(figsize=(6, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()],
         'r--', label='Perfect Prediction')
plt.xlabel('Actual Volatility')
plt.ylabel('Predicted Volatility')
plt.title('Prediction Accuracy')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

print("\n" + "=" * 50)
print("PROJECT COMPLETE")
print("=" * 50)
```
